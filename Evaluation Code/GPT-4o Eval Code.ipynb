```python
# SST-2 Eval

import openai
import pandas as pd

# Set your OpenAI API key
openai.api_key = ""  # Replace with your actual OpenAI API key

# Load the evaluation CSV file
evaluation_csv_path = '/content/drive/MyDrive/Algoverse/New Results/GPT 4o mini Translations/Translated SST-2.csv'
evaluation_data = pd.read_csv(evaluation_csv_path)

# Define a function to determine sentiment
def determine_sentiment(sentence):
    prompt = (
        "You are a sentiment analysis expert. Determine the sentiment of the following sentence. "
        "If the sentence expresses a positive sentiment, return '1'. If it expresses a negative sentiment, return '0'. "
        "Focus only on identifying whether the overall sentiment is positive or negative. Only output            0 or 1, nothing else.\n\n"
        f"Sentence: {sentence}\n\n"
        "Sentiment (0 = negative, 1 = positive):"
    )

    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "You are a helpful assistant that performs sentiment analysis."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=10,
            temperature=0
        )

        # Extract the sentiment value
        sentiment = response.choices[0].message['content'].strip()

        # Ensure the sentiment is a valid integer (0 or 1)
        if sentiment in ['0', '1']:
            return int(sentiment)
        else:
            print(f"Unexpected response for sentence '{sentence}': {sentiment}")
            return 0  # Default to 0 if the response is unexpected
    except Exception as e:
        print(f"Error processing sentence '{sentence}': {e}")
        return 0  # Default to 0 in case of error

# Process each sentence and calculate SAE and AAVE Answers
sae_answers = []
aave_answers = []
for index, row in evaluation_data.iterrows():
    original_sentence = row['Original Sentence']
    translated_sentence = row['Translated Sentence']

    # Determine sentiments
    sae_answer = determine_sentiment(original_sentence)
    aave_answer = determine_sentiment(translated_sentence)

    sae_answers.append(sae_answer)
    aave_answers.append(aave_answer)

    print(f"Processed row {index + 1}: SAE Sentiment {sae_answer}, AAVE Sentiment {aave_answer}")

# Add the SAE and AAVE Answer columns to the DataFrame
evaluation_data['SAE Answer'] = sae_answers
evaluation_data['AAVE Answer'] = aave_answers

# Reorder columns to place SAE Answer after Actual Answer and before AAVE Answer
evaluation_data = evaluation_data[['Original Sentence', 'Translated Sentence', 'Actual Label', 'SAE Answer', 'AAVE Answer']]

# Save the updated results to a new CSV file
new_csv_path = '/content/drive/MyDrive/Algoverse/New Results/Evaluation Results/GPT-4o Eval Results/SST-2/SST-2_Evaluation_GPT-4o.csv'
evaluation_data.to_csv(new_csv_path, index=False)

print(f"SAE and AAVE evaluation completed and saved to {new_csv_path}")
```

```python
## BoolQ Evaluation Code

import openai
import pandas as pd
import re

# Set your OpenAI API key
openai.api_key = ""  # Replace with your actual OpenAI API key

# Load the BoolQ translation CSV file
boolq_csv_path = '/content/drive/MyDrive/Algoverse/New Results/GPT 4o mini Translations/Translated BoolQ.csv'
boolq_data = pd.read_csv(boolq_csv_path)

# Define a function to determine the answer (TRUE/FALSE) using OpenAI
def determine_true_false(passage, question):
    prompt = (
        "You are a text analysis expert. Determine if the following passage supports answering the question as TRUE or FALSE. "
        "Return 'TRUE' or 'FALSE' in all caps. Only output TRUE or FALSE, nothing else. \n\n"
        f"Passage: {passage}\n"
        f"Question: {question}\n\n"
        "Answer (TRUE or FALSE):"
    )

    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "You are a helpful assistant that evaluates true/false questions."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=10,
            temperature=0
        )

        # Extract the answer
        answer = response.choices[0].message['content'].strip()

        # Use regex to ensure the answer is a single word, TRUE or FALSE
        if re.fullmatch(r'\b(TRUE|FALSE)\b', answer, re.IGNORECASE):
            return answer.upper()
        else:
            print(f"Unexpected response for passage '{passage}' and question '{question}': {answer}")
            return "FALSE"  # Default to FALSE if the response is unexpected
    except Exception as e:
        print(f"Error processing passage '{passage}' and question '{question}': {e}")
        return "FALSE"  # Default to FALSE in case of error

# Process each passage and question to calculate SAE and AAVE answers
sae_answers = []
aave_answers = []
for index, row in boolq_data.iterrows():
    sae_passage = row['SAE Passage']
    sae_question = row['SAE Question']
    aave_passage = row['AAVE Passage']
    aave_question = row['AAVE Question']

    # Determine TRUE/FALSE for SAE
    sae_answer = determine_true_false(sae_passage, sae_question)
    sae_answers.append(sae_answer)

    # Determine TRUE/FALSE for AAVE
    aave_answer = determine_true_false(aave_passage, aave_question)
    aave_answers.append(aave_answer)

    print(f"Processed row {index + 1}: SAE Answer {sae_answer}, AAVE Answer {aave_answer}")

# Add the SAE and AAVE Answer columns to the DataFrame
boolq_data['SAE Answer'] = sae_answers
boolq_data['AAVE Answer'] = aave_answers

# Reorder columns to place SAE and AAVE Answer after Original Label
boolq_data = boolq_data[['SAE Passage', 'SAE Question', 'AAVE Passage', 'AAVE Question', 'Actual Label', 'SAE Answer', 'AAVE Answer']]

# Save the updated results to a new CSV file
updated_csv_path = '/content/drive/MyDrive/Algoverse/New Results/Evaluation Results/GPT-4o Eval Results/BoolQ/BoolQ_Evaluation_GPT-4o.csv'
boolq_data.to_csv(updated_csv_path, index=False)

print(f"SAE and AAVE evaluation completed and saved to {updated_csv_path}")
```

```python
# MultiRC Evaluation Code

import openai
import pandas as pd
import re

# Set your OpenAI API key
openai.api_key = ""  # Replace with your actual OpenAI API key

# Load the MultiRC translation CSV file
multirc_csv_path = '/content/drive/MyDrive/Algoverse/New Results/GPT 4o mini Translations/Translated MultiRC.csv'
multirc_data = pd.read_csv(multirc_csv_path)

# Define a function to determine the truth value (TRUE/FALSE) using OpenAI
def determine_true_false(paragraph, question, answer_choice):
    prompt = (
        "You are a text analysis expert. Determine if the following answer choice is correct for the given paragraph and question. "
        "Return '1' for true and '0' for false. Only output 0 or 1, nothing else.\n\n"
        f"Paragraph: {paragraph}\n"
        f"Question: {question}\n"
        f"Answer Choice: {answer_choice}\n\n"
        "Answer (0 = false, 1 = true):"
    )

    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "You are a helpful assistant that evaluates true/false questions."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=10,
            temperature=0
        )

        # Extract the answer
        answer = response.choices[0].message['content'].strip()

        # Use regex to ensure the answer is a single digit, 0 or 1
        if re.fullmatch(r'\b[01]\b', answer):
            return int(answer)
        else:
            print(f"Unexpected response for paragraph '{paragraph}' and question '{question}' with answer choice '{answer_choice}': {answer}")
            return 0  # Default to 0 if the response is unexpected
    except Exception as e:
        print(f"Error processing paragraph '{paragraph}', question '{question}', and answer choice '{answer_choice}': {e}")
        return 0  # Default to 0 in case of error

# Process each paragraph, question, and answer choice to calculate SAE and AAVE answers
sae_answers = []
aave_answers = []
for index, row in multirc_data.iterrows():
    sae_paragraph = row['Paragraph']
    sae_question = row['Question']
    sae_answer_choice = row['Answer Choice']
    aave_paragraph = row['Translated Paragraph']
    aave_question = row['Translated Question']
    aave_answer_choice = row['Translated Answer Choice']

    # Determine true/false for SAE
    sae_answer = determine_true_false(sae_paragraph, sae_question, sae_answer_choice)
    sae_answers.append(sae_answer)

    # Determine true/false for AAVE
    aave_answer = determine_true_false(aave_paragraph, aave_question, aave_answer_choice)
    aave_answers.append(aave_answer)

    print(f"Processed row {index + 1}: SAE Answer {sae_answer}, AAVE Answer {aave_answer}")

# Add the SAE and AAVE Answer columns to the DataFrame
multirc_data['SAE Answer'] = sae_answers
multirc_data['AAVE Answer'] = aave_answers

# Rename 'Actual Label' to 'Actual Answer'
multirc_data.rename(columns={'Actual Label': 'Actual Answer'}, inplace=True)

# Reorder columns as specified
multirc_data = multirc_data[['Paragraph', 'Question', 'Answer Choice', 'Translated Paragraph', 'Translated Question', 'Translated Answer Choice', 'Actual Answer', 'SAE Answer', 'AAVE Answer']]

# Save the updated results to a new CSV file
updated_csv_path = '/content/drive/MyDrive/Algoverse/New Results/Evaluation Results/GPT-4o Eval Results/MultiRC/MultiRC_Evaluation_GPT4o.csv'
multirc_data.to_csv(updated_csv_path, index=False)

print(f"SAE and AAVE evaluation completed and saved to {updated_csv_path}")
```

```python
# Copa Evaluation Code

# Copa Evaluation Code

import openai
import pandas as pd
import re

# Set your OpenAI API key
openai.api_key = ""  # Replace with your actual OpenAI API key

# Load the COPA translation CSV file
copa_csv_path = '/content/drive/MyDrive/Algoverse/New Results/GPT 4o mini Translations/Translated Copa.csv'
copa_data = pd.read_csv(copa_csv_path)

# Define a function to determine the correct choice using OpenAI
def determine_choice(premise, choice1, choice2):
    prompt = (
        "You are a text analysis expert. Determine which choice more plausibly has a causal relationship with the given premise. "
        "Return '0' if the first choice is correct and '1' if the second choice is correct. Only output 0 or 1, nothing else.\n\n"
        f"Premise: {premise}\n"
        f"Choice 1: {choice1}\n"
        f"Choice 2: {choice2}\n\n"
        "Answer (0 = Choice 1, 1 = Choice 2):"
    )

    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "You are a helpful assistant that evaluates causal relationships."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=10,
            temperature=0
        )

        # Extract the answer
        answer = response.choices[0].message['content'].strip()

        # Use regex to ensure the answer is a single digit, 0 or 1
        if re.fullmatch(r'\b[01]\b', answer):
            return int(answer)
        else:
            print(f"Unexpected response for premise '{premise}' with choices '{choice1}' and '{choice2}': {answer}")
            return 0  # Default to 0 if the response is unexpected
    except Exception as e:
        print(f"Error processing premise '{premise}', choice 1 '{choice1}', and choice 2 '{choice2}': {e}")
        return 0  # Default to 0 in case of error

# Process each premise and choices to calculate SAE and AAVE answers
sae_answers = []
aave_answers = []
for index, row in copa_data.iterrows():
    sae_premise = row['Premise']
    sae_choice1 = row['Choice 1']
    sae_choice2 = row['Choice 2']
    aave_premise = row['Translated Premise']
    aave_choice1 = row['Translated Choice 1']
    aave_choice2 = row['Translated Choice 2']

    # Determine the correct choice for SAE
    sae_answer = determine_choice(sae_premise, sae_choice1, sae_choice2)
    sae_answers.append(sae_answer)

    # Determine the correct choice for AAVE
    aave_answer = determine_choice(aave_premise, aave_choice1, aave_choice2)
    aave_answers.append(aave_answer)

    print(f"Processed row {index + 1}: SAE Answer {sae_answer}, AAVE Answer {aave_answer}")

# Add the SAE and AAVE Answer columns to the DataFrame
copa_data['SAE Answer'] = sae_answers
copa_data['AAVE Answer'] = aave_answers

# Reorder columns as specified
copa_data = copa_data[['Premise', 'Choice 1', 'Choice 2', 'Translated Premise', 'Translated Choice 1', 'Translated Choice 2', 'Actual Answer', 'SAE Answer', 'AAVE Answer']]

# Save the updated results to a new CSV file
updated_csv_path = '/content/drive/MyDrive/Algoverse/New Results/Evaluation Results/GPT-4o Eval Results/COPA/COPA_Evaluation_GPT4o.csv'
copa_data.to_csv(updated_csv_path, index=False)

print(f"SAE and AAVE evaluation completed and saved to {updated_csv_path}")
```

```python
# WSC

import openai
import pandas as pd
import re

# Set your OpenAI API key
openai.api_key = ""  # Replace with your actual OpenAI API key

# Load the WSC CSV file
wsc_csv_path = '/content/drive/MyDrive/Algoverse/New Results/GPT 4o mini Translations/Translated WSC.csv'
wsc_data = pd.read_csv(wsc_csv_path)

# Define a function to determine coreference
def determine_coreference(paragraph, span1, span2):
    prompt = (
        "You are reading a passage and are solving a coreference resolution task in which examples consist of a sentence with a pronoun and a list of two noun phrases from the sentence. "
        "You must determine the correct referent of the pronoun from among the provided choices. "
        "Only respond with '1' if Span 2 refers to Span 1, or '0' if Span 2 does not. "
        "You can only respond with one of these two things: '0' or '1'.\n\n"
        f"Paragraph: {paragraph}\n"
        f"Span 1: {span1}\n"
        f"Span 2: {span2}\n\n"
        "Coreference Resolution (0 = Span 2 does not refer to Span 1, 1 = Span 2 refers to Span 1):"
    )

    try:
        response = openai.ChatCompletion.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "You are a helpful assistant that resolves coreference tasks."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=10,
            temperature=0
        )

        # Extract the answer
        answer = response.choices[0].message['content'].strip()

        # Use regex to ensure the answer is a single digit, 0 or 1
        if re.fullmatch(r'\b[01]\b', answer):
            return int(answer)
        else:
            print(f"Unexpected response for paragraph '{paragraph}' with spans '{span1}' and '{span2}': {answer}")
            return 0  # Default to 0 if the response is unexpected
    except Exception as e:
        print(f"Error processing paragraph '{paragraph}', span 1 '{span1}', and span 2 '{span2}': {e}")
        return 0  # Default to 0 in case of error

# Process each paragraph to calculate SAE and AAVE Answers
sae_answers = []
aave_answers = []
for index, row in wsc_data.iterrows():
    original_paragraph = row['Original Paragraph']
    aave_paragraph = row['AAVE Paragraph']
    span1 = row['Span 1']
    span2 = row['Span 2']

    # Determine coreferences
    sae_answer = determine_coreference(original_paragraph, span1, span2)
    aave_answer = determine_coreference(aave_paragraph, span1, span2)

    sae_answers.append(sae_answer)
    aave_answers.append(aave_answer)

    print(f"Processed row {index + 1}: SAE Coreference {sae_answer}, AAVE Coreference {aave_answer}")

# Add the SAE and AAVE Answer columns to the DataFrame
wsc_data['SAE Answer'] = sae_answers
wsc_data['AAVE Answer'] = aave_answers

# Reorder columns to place SAE Answer after Actual Label and before AAVE Answer
wsc_data = wsc_data[['Original Paragraph', 'AAVE Paragraph', 'Span 1', 'Span 2', 'Actual Label', 'SAE Answer', 'AAVE Answer']]

# Save the updated results to a new CSV file
new_csv_path = '/content/drive/MyDrive/Algoverse/New Results/Evaluation Results/GPT-4-turbo Eval Results/WSC/WSC_Evaluation_GPT4Turbo.csv'
wsc_data.to_csv(new_csv_path, index=False)

print(f"SAE and AAVE evaluation completed and saved to {new_csv_path}")
```
