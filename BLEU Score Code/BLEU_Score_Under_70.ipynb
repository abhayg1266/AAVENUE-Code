```python
import pandas as pd
import gdown
from google.colab import drive
drive.mount('/content/drive')
```

```python
sst2_id = "10Jccd_BHBaPG_oP_6nV7tjiEVw-riTFA"
boolq_id = "10XkGCVcIFZg8nRT2XldyEphc3nj6igkr"
copa_id = "10X5fqEGp2S49H-3-bdxLg06TNqx-66kW"
wsc_id = "10SfFQA_u2DWcB7lV6O3db7UU3E7oDlbE"
multirc_id = "10OACT7pGN0C6yZ0SINFv0GaayCI6syCB"

url1 = f'https://drive.google.com/uc?id={sst2_id}'
url2 = f'https://drive.google.com/uc?id={boolq_id}'
url3 = f'https://drive.google.com/uc?id={copa_id}'
url4 = f'https://drive.google.com/uc?id={wsc_id}'
url5 = f'https://drive.google.com/uc?id={multirc_id}'

gdown.download(url1, 'file1.csv', quiet=False)
gdown.download(url2, 'file2.csv', quiet=False)
gdown.download(url3, 'file3.csv', quiet=False)
gdown.download(url4, 'file4.csv', quiet=False)
gdown.download(url5, 'file5.csv', quiet=False)

sst2 = pd.read_csv("file1.csv")
boolq = pd.read_csv("file2.csv")
copa = pd.read_csv("file3.csv")
wsc = pd.read_csv("file4.csv")
multirc = pd.read_csv("file5.csv")
```

```python
filtered_sst2 = sst2[sst2['BLEU Score Sentence'] < 0.7]

output_results_csv = '/content/drive/MyDrive/New Results/BLEU Under 70 Evals/Gemini-1.5-Pro/SST-2/sst2_under_70.csv'
filtered_sst2.to_csv(output_results_csv, index=False)
print(f"Results with BLEU scores saved to '{output_results_csv}'.")
```

```python
filtered_boolq = boolq[(boolq['BLEU Score Passage'] < 0.7) & (boolq['BLEU Score Question'] < 0.7)]

output_results_csv = '/content/drive/MyDrive/New Results/BLEU Under 70 Evals/Gemini-1.5-Pro/BoolQ/boolq_under_70.csv'
filtered_boolq.to_csv(output_results_csv, index=False)
print(f"Results with BLEU scores saved to '{output_results_csv}'.")
```

```python
filtered_copa = copa[(copa['BLEU Score Premise'] < 0.7) & (copa['BLEU Score Choice 1'] < 0.7) & (copa['BLEU Score Choice 2'] < 0.7)]

output_results_csv = '/content/drive/MyDrive/New Results/BLEU Under 70 Evals/Gemini-1.5-Pro/COPA/copa_under_70.csv'
filtered_copa.to_csv(output_results_csv, index=False)
print(f"Results with BLEU scores saved to '{output_results_csv}'.")
```

```python
filtered_wsc = wsc[(wsc['BLEU Score Paragraph'] < 0.7)]

output_results_csv = '/content/drive/MyDrive/New Results/BLEU Under 70 Evals/Gemini-1.5-Pro/WSC/wsc_under_70.csv'
filtered_wsc.to_csv(output_results_csv, index=False)
print(f"Results with BLEU scores saved to '{output_results_csv}'.")
```

```python
filtered_multirc = multirc[(multirc['BLEU Score Paragraph'] < 0.7) & (multirc['BLEU Score Question'] < 0.7) & (multirc['BLEU Score Answer Choice'] < 0.7)]

output_results_csv = '/content/drive/MyDrive/New Results/BLEU Under 70 Evals/Gemini-1.5-Pro/MultiRC/multirc_under_70.csv'
filtered_multirc.to_csv(output_results_csv, index=False)
print(f"Results with BLEU scores saved to '{output_results_csv}'.")
```
